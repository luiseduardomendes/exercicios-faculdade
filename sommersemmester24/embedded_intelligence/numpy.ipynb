```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(1)

array = (np.random.rand(100) - 0.5) * 10
#array = np.sin(np.linspace(0, 10*np.pi, 100))
fft_array = np.fft.fft(array)
fft_array_real = fft_array.real
fft_array_imag = fft_array.imag
fft_array_norm = np.sqrt(fft_array_real ** 2 + fft_array_imag ** 2)
fft_array_angl = np.angle(fft_array)

df = pd.DataFrame({
  "fft": fft_array,
  "real": fft_array_real,
  "imag": fft_array_imag,
  "norm": fft_array_norm,
  "angl": fft_array_angl,
})

plt.plot(fft_array_angl)
plt.grid()
plt.show()

# df.head(5)
fft_array_real.mean()
```
