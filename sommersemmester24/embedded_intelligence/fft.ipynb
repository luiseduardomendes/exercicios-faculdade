```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint

N = 32
t = np.linspace(0, 2*np.pi, N)

m = 4.
nu = m/N
#f = np.sin(2*np.pi*nu*t) + np.sin(3*2*np.pi*nu*t) / 3
f = 0
print("f =", end='')
for n in range(1, 100, 2):
  print(f" + sin({n}t)/{n}", end='')
  f += np.sin(n*t)/n
print()

ft = np.fft.fft(f)
freq = np.fft.fftfreq(N)
amp = ft.real**2 + ft.imag**2

freq_amp = np.column_stack((freq, amp))
sorted_freq_amp = freq_amp[freq_amp[:, 0].argsort()]
sorted_freq = sorted_freq_amp[:, 0]
sorted_amp = sorted_freq_amp[:, 1]
dt = np.column_stack((sorted_freq, sorted_amp))

#pprint(dt)
#plt.plot(freq, amp)
plt.plot(sorted_freq, sorted_amp)
#plt.plot(t, f)
plt.show()
```
