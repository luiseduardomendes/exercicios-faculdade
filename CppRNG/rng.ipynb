```python
import numpy as np
import matplotlib.pyplot as plt


t1 = np.linspace(0, 0.01, 100) 
t2 = np.linspace(0.01, 0.1, 900)

v1 = 80 * np.exp(-40*t1)
v2 = 80*np.exp(0.1) * np.exp(-50*t2)

t = np.hstack([t1, t2])
v = np.hstack([v1, v2])


plt.plot(t, v)
plt.scatter([0,0.01,0.1], [80 * np.exp(-40*0), 80 * np.exp(-40*0.01), 80*np.exp(0.1) * np.exp(-50*0.1)])
plt.grid()
plt.show()
```
