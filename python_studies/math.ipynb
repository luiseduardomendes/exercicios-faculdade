```python
import math
import time
for n in range(1,1000):
    a = n
    ck_begin = time.time()
    while a > 1:
        ck_curr = time.time()
        digits = [int(i) for i in str(a)]
        #print(digits)
        digits = [i**2 for i in digits]
        #print(digits)
        sum = 0
        for digit in digits:
            sum += digit
        a = sum
        #print(a)
        if a > 9999 or ck_curr - ck_begin > 0.05:
            break
    if a == 1:
        print(n)
```
