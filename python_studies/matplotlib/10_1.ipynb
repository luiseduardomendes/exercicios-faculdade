```python
import matplotlib.pyplot as plt
import numpy as np

xa_array = np.linspace(0,2*np.pi,100)
ya_array = np.sin(xa_array)

xb_array = np.linspace(0,2*np.pi,100)
yb_array = np.cos(xb_array)

# insert the data on the graph
plt.plot(xa_array, ya_array)
plt.plot(xb_array, yb_array)

# rename the axis
plt.xlabel('x')
plt.ylabel('y')

# plot the grid
plt.grid()
```

```python
fig, ax = plt.subplots()

ax.plot(xa_array,ya_array,'r--')
ax.plot(xb_array,yb_array,'b--')

ax.grid()

ax.set(
    xlabel='x',
    ylabel='y',
    title='Sine and Cosine functions',
)
```

## Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xa_array = np.linspace(0,np.e,100)
ya_array = np.exp(xa_array)
yb_array = np.log(xa_array)
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(xa_array, ya_array)
ax[0][1].plot(xa_array, yb_array)
ax[0][0].grid()
ax[0][1].grid()
```

```python
plt.figure()

xa_array = np.linspace(0,2*np.pi,100)
plt.subplot(211)
plt.plot(xa_array, np.sin(xa_array))
plt.plot(xa_array, np.cos(xa_array))
plt.legend(['sine function', 'cosine function'])
plt.grid()

xb_array = np.linspace(-2*np.e,2*np.e,100)
plt.subplot(212)
plt.plot(xb_array, np.sinh(xb_array))
plt.plot(xb_array, np.cosh(xb_array))
plt.legend(['hyperbolic sine function', 'hyperbolic cosine function'])
plt.grid()
```

## Logarithmic scale

```python
import matplotlib.pyplot as plt
import numpy as np

xa_array = np.linspace(0,10,500)
ya_array = np.sinh(xa_array)

fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(xa_array, ya_array)
ax1.set_yscale('log')
ax1.grid()

ax2.plot(xa_array, ya_array)
ax2.grid()
```

## Secondary axis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xa_array = np.linspace(0,10,12)
ya_array = np.sin(xa_array)+5
yb_array = 10-np.exp(-xa_array)

fig, ax1 = plt.subplots()
ax1.bar(xa_array,ya_array)
ax1.set(
    title='',
    xlabel='x1',    
) 
ax1.set_ylabel(ylabel='y1',color='b')
ax1.grid(color='b')

ax2 = ax1.twinx()

ax2.plot(xa_array,yb_array,color='r')
ax2.set_ylabel('y2', color='r')

ax2.grid(color='r')

```
