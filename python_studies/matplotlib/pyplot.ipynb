```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,25, dtype='float')
y_sin = np.sin(x)

y_cos = np.cos(x)

sigma = np.sqrt(np.exp(-1))
y_gauss = np.exp(-0.5*(x/sigma)**2) / (sigma * np.sqrt(2*np.pi))

plt.plot(x, y_sin)
plt.plot(x, y_cos, '--')
plt.plot(x, y_gauss, 'ro')


plt.xlabel('x axis')
plt.ylabel('sin(x)')

plt.grid()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y_gauss, '-', label='Gaussian Distribution')

ax.set(xlabel='x axis', ylabel='y axis')
ax.grid()

ax.set_title('Gaussian Distribution (sigma^2 = e^-1)')
ax.legend()

```
