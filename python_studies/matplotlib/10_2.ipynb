```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi,np.pi,100)
y = np.linspace(-np.pi,np.pi,100)
x,y = np.meshgrid(x, y)
z = np.sin(x) * np.sin(y) 
```

```python
fig = plt.figure()
ax1 = fig.gca(projection='3d')
ax1.plot_surface(x, y, z)
plt.title('Image plot of $sin(x) \cdot sin(y)$ for a grid of values')
plt.show()
plt.close()
plt.title('Image plot of $sin(x) \cdot sin(y)$ for a grid of values')
plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
```
