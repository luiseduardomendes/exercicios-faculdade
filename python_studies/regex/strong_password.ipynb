```python
#strong password generator
from random import randint

passwords_list = []

for j in range(1000):
    string = '' 
    for i in range(randint(10, 20)):
        random_value = chr(randint(32,126))
        string+=random_value
    passwords_list.append(string)

passwords_list
```

```python
import re

pattern_password = re.compile(
    r'''
    ^
    (?!.*[ ]) # has no spaces
    (?=.*[a-z]) # has a character lower
    (?=.*[A-Z]) # has a character upper
    (?=.*[0-9]) # has a digit
    (?=.*[!-/\:-@\[-`\|-~]) # has a different character
    .{12,} # has more than 12 characters
    $
    ''',
    flags=re.VERBOSE
)

strong_passwords = [pattern_password.findall(password) for password in passwords_list if len(pattern_password.findall(password)) > 0]

strong_passwords
```
