## Validação de IPv4

[0-255].[0-255].[0-255].[0-255]
exemplo = 128.251.0.12

```python
cpf_example = '025.258.963-10'
ipv4_example = '128.251.0.12'
```

```python
import re

cpf_pattern = re.compile(r'^\d{3}\.\d{3}\.\d{3}\-\d{2}$')


ipv4_pattern = re.compile(r'''
^
(?:25[0-5] | 2[0-4][0-9] | 1[0-9][0-9] | [1-9][0-9] | [0-9])
\.
(?:25[0-5] | 2[0-4][0-9] | 1[0-9][0-9] | [1-9][0-9] | [0-9])
\.
(?:25[0-5] | 2[0-4][0-9] | 1[0-9][0-9] | [1-9][0-9] | [0-9])
\.
(?:25[0-5] | 2[0-4][0-9] | 1[0-9][0-9] | [1-9][0-9] | [0-9])
$
'''
,flags=re.X)
```

```python
cpf_pattern.findall(cpf_example)
```

```python
ipv4_pattern.findall(ipv4_example)
```

```python
from random import randint

ip_list = []

for i in range(0,1000):
    ip_list.append(f'{randint(0,500)}.{randint(0,500)}.{randint(0,500)}.{randint(0,500)}')
ip_list
```

```python
valid_ip_list = [ipv4_pattern.findall(ip) for ip in ip_list if len(ipv4_pattern.findall(ip)) != 0]
valid_ip_list
```
