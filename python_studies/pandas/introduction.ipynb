4. estado de onde sai mais jogadores

```python
states = [state for state in df['birth_state'] if state != 'NaN' and type(state) == str]
states = pd.Series(list(set(states)))

players_by_state = pd.Series(np.zeros(len(states)))

for i, state in enumerate(df['birth_state']):
    if state != 'NaN' and type(state) == str:
        index = list(states).index(state)
        players_by_state[index] += 1

df2 = pd.DataFrame(
    {
        'states': states,
        'num_players': players_by_state
    }
)

i = df2['num_players'].idxmax()
df2['states'][i]
```
