```python
import pandas as pd
import numpy as np
```

```python
path_output_file = 'data/random_values_5x3.csv'

n_rows = 5
n_cols = 5

inf_rng = 0
sup_rng = 15

random_values = [np.random.randint(inf_rng, sup_rng) for i in range(n_rows * n_cols)]
#random_values.sort()
random_values = np.array(random_values).reshape(n_rows, n_cols)

columns_names = np.array(['column ' + str(i+1) for i in range(n_cols)])

df = pd.DataFrame(random_values, columns=columns_names)

df.to_csv(path_output_file, index=False)
```

```python
df = pd.read_csv(path_output_file)
df.plot.bar()
```

```python
a = np.linspace(-np.pi, np.pi, 100)

df = pd.DataFrame({
    'index': a,
    'cos(x)': np.cos(a),
    'sin(x)': np.sin(a)
})

df.plot(x='index', y='cos(x)')
```
