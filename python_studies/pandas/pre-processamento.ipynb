            # If ‘ignore’, suppress error and only existing labels are dropped.

```python
df = pd.read_csv('data/random_values_5x3.csv')
df.drop(
    index=np.arange(2,4),
    columns=df.columns[0:2]
)
```

```python
df.drop(
    labels='column 1',
    axis=1
)
```
