```python
import pandas as pd
import numpy as np
```

```python
file_path = 'data/horarios.csv'

df = pd.read_csv(file_path, sep=';', encoding='latin')
df.columns
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['horario'], dayfirst=False)
df.head(3)
```

```python
filter1 = df['razao_social'].value_counts() == 2
filtered_list = list(filter1[filter1].index)

filtered_df = df[df['razao_social'].isin(filtered_list)]
```

```python
series_min = pd.pivot_table(filtered_df, index='razao_social', values='time', aggfunc='min')
series_max = pd.pivot_table(filtered_df, index='razao_social', values='time', aggfunc='max')

series_max - series_min
```
