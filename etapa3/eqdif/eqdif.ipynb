```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

n, x = sp.symbols(list('nx'))

#fnx = (((-1)**n)/(n**2 * np.pi**2)) * sp.cos((n * np.pi * x) / 2) + ((4 * (-1)**(n+1))/((2*n - 1)**2 * np.pi**2)) * sp.sin((n * np.pi * x) / 2)
#fnx = ((((-2 * sp.cos(n * np.pi / 2))/(n*np.pi)) + ((4 * sp.sin(n * np.pi / 2))/(n**2 * np.pi ** 2))) * sp.sin(n * np.pi * x / 2))
#fnx = (((2*sp.cos((2*n - 1) * x))/(sp.pi*(2*n - 1)**2)) + (((-1) ** (n-1) * sp.sin(n * x))/n))
#fnx = (((4* sp.pi * (-1) ** n) / (n ** 2) * sp.cos(n*x)))
fnx = ((-8 (2 + (-2 + n**2 * sp.pi**2) (-1)**n)/n**3)) * sp.sin(n/2 * x)
fnx
```

```python
def plot(n_max):
    s = 0
    for i in range(1, n_max):
        s += fnx.subs(n, i)
    s += 2 * sp.pi ** 3 / 3

    x_axis = np.linspace(-10, 10, 500)
    y_axis = []
    for i in x_axis:
        y_axis.append(s.subs(x, i))

    plt.plot(x_axis, y_axis)
```

```python
plot(10)
```
