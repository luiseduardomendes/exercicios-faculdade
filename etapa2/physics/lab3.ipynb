```python
import numpy as np
import sympy as sp
import pandas as pd
import matplotlib.pyplot as plt

sp.init_printing()
```

```python
df = pd.read_csv('lab3_data.csv')
data = df.to_dict('list') 
x = np.array(data['x'])
y = np.array(data['y'])

_A = 1/(x**2)
A = sp.Matrix(_A)

x1 = sp.symbols('x1')
var = sp.linsolve((A.T*A, A.T*sp.Matrix(y)), )

plt.scatter(x, y)

n1 = A.T*A
n2 = A.T*sp.Matrix(y)

beta = n2.det()/n1.det()

_x = np.linspace(x[0], x[-1], 100)
_y = beta/(_x**2)

plt.plot(_x, _y)
plt.show()
```
