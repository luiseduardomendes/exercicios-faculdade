```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp.init_printing()
```

```python
y = np.array([55, 72, 100, 140, 206, 333, 582, 1111, 2601, 7970])
x = np.array([0.1, 0.09, 0.08, 0.07, 0.06, 0.05, 0.04, 0.03, 0.02, 0.01])

x1 = np.linspace(0.01,0.1, 100)
y1 = 0.416*2/(x1**2)

fig, ax = plt.subplots()

ax.plot(x1, y1, 'r', label='Ajuste')
ax.scatter(x, y, label='Dados obtidos')
ax.set(xlabel='distância(m)', ylabel='Campo Magnético(μT)')
ax.grid()
ax.legend()
ax.set_title('Variação do campo magnético em função da distância')

fig.savefig('MagFieldGraph.png')
```
